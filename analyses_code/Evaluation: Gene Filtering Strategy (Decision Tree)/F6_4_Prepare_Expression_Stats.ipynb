```r
## Prepare the expression based summary stats for decision tree analysis
```

# Libraries

```r
source('MS0_Libraries.r')
```

```r
library(stringi)
```

# Parameters

```r
### Define dataset and cell-type for which to prepare
```

```r
### dataset
```

```r
dataset = 'oneK1k'
```

```r
dir = 'oneK1K'
```

```r
cell_type = 'CD8_T'
```

```r
### Define which summary stats to use
```

```r
type = 'data' # data or  'RNA' --> in case of data normalized expression stats are used
```

```r
### Path to the summary stats
```

```r
data_path_summary_stats = '../data/current/summary_stats/'

## Files: Sample_gene_statistic  (info about genes on sample level)
#         Sample_summary (information about samples)
#         Sample_cell statistic
```

```r
### Path to save prepared summary stats to
```

```r
output_path = paste0('../data/current/coeqtl_mapping/co_qtls_sceqtlgen/analysis_',dir , '/', cell_type, '/')

## co_qtls_sceqtlgen can be replaced by co_qtls_decision_tree
## co_qtls_sceqtlgen: contains all for initial mapping run
## co_qtls_decision_tree: files for final decision tree filtered run
```

```r
### Summary stats to load based on parameters
```

```r
if(type == 'data'){
    summary_stats_name = paste0("D1_Sample_gene_statisticwg3_", dataset , cell_type,  ".Qced.Normalized.SCs.Rds_data.csv") ## test with normalized stats
    }

if(type == 'RNA'){
    summary_stats_name = paste0("D1_Sample_gene_statisticwg3_", dataset, cell_type,  ".Qced.Normalized.SCs.Rds.csv")  # !! adjust cell-type
}
```

# Data

```r
## Load expression summary stats
```

```r
gene_sample_stats = read.csv(paste0(data_path_summary_stats, summary_stats_name))
```

```r
paste0(data_path_summary_stats, summary_stats_name)
```

```r
gene_sample_stats = setDT(gene_sample_stats)
```

```r
head(gene_sample_stats,2)
```

```r
### Aggregate per gene to map to gene-pairs
```

```r
gene_summary_stats = gene_sample_stats %>% group_by(gene) %>% summarise(
    mean_expression = mean(mean_counts),
    sd_sample_mean_expression = sd(mean_counts),
    mean_variance = mean(variance_counts),
    sd_sample_variance = sd(variance_counts),
    sum_non_zero = sum(amount_non_zero_counts),
    mean_non_zero = mean(amount_non_zero_counts),
    mean_percentage_zero = mean(percentage_zero),
    sum_zero = sum(amount_zero_counts),
    mean_zero = mean(amount_zero_counts),
    mean_coef_var = mean(sqrt(variance_counts)/ mean_counts)
    )
    
```

```r
head(gene_summary_stats,2)
```

```r
### Save prepared summary stats
```

```r
if(type == 'data'){
    write.csv(gene_summary_stats, paste0(output_path, 'F6_Expression_Summary_Stats_all_data.csv')) # calculated based on normalized values
    }

if(type == 'RNA'){
    write.csv(gene_summary_stats, paste0(output_path, 'F6_Expression_Summary_Stats_all.csv'))
    }
```

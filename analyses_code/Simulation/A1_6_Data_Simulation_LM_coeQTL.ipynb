## Linear model on z-value

```r
## Estimate the linear model on the z-value transformed correlations instead of the correlations
```

```r
### Define how to handle infinite z-value
# 1) Set to NA: ''
# 2) Replace by max: 'max'
```

```r
z_value_handle = ''
```

```r
if(z_value_handle == ''){
    cor_data_long$z_value[is.infinite(cor_data_long$z_value)] = NA
    cor_data_long$z_value_non_zero[is.infinite(cor_data_long$z_value_non_zero)] = NA
    }

if(z_value_handle == 'max'){
    cor_data_long$z_value[is.infinite(cor_data_long$z_value)]  = max(cor_data_long$z_value[!is.infinite(cor_data_long$z_value)])
    cor_data_long$z_value_non_zero[is.infinite(cor_data_long$z_value_non_zero)]  = max(cor_data_long$z_value[!is.infinite(cor_data_long$z_value_non_zero)])
    }
    
```
